```python
import json

# Læs linjer og gør dem til en liste af dicts
with open("/Users/jeppekoch/Projects/BI/house-price.json") as f:
    lines = f.readlines()
    data = [json.loads(line.strip()) for line in lines]

#Brug pandas
import pandas as pd
df = pd.json_normalize(data)

print(df.head())
```

```python
df = df.dropna()
df
```

```python
df_categorical2 = df.select_dtypes(exclude=[np.number])

df_categorical2
```

```python


df_categorical2['mainroad'] = df_categorical2['mainroad'].replace({"no":0, "yes":1})
df_categorical2['guestroom'] = df_categorical2['guestroom'].replace({"no":0,"yes":1})
df_categorical2['basement'] = df_categorical2['basement'].replace({"no":0, "yes":1})
df_categorical2['hotwaterheating'] = df_categorical2['hotwaterheating'].replace({"no":0, "yes":1})
df_categorical2['airconditioning'] = df_categorical2['airconditioning'].replace({"no":0, "yes":1})
df_categorical2['prefarea'] = df_categorical2['prefarea'].replace({"no":0, "yes":1})
df_categorical2['furnishingstatus'] = df_categorical2['furnishingstatus'].replace({"furnished":1, "semi-furnished":2, "unfirnished":3})
```
