```python
import pandas as pd
import matplotlib.pyplot as plt

# Loaded the local file into pandas DataFrame
df = pd.read_excel(r"C:\Users\benja\BI Skole\Projects\SuperStoreUS-2015.xlsx")

# Checked if there were any data missing in cells
print('')
print(df.isna().sum())

# removed the rows with NULL values, the dataset was big enough that i could remove 16rows
df_clean = df.dropna()
# Checked if everything looks as expected
print(df_clean.isna().sum())

# Looked at column names to see if there were any sensitive data
df_clean.columns
# Removed the Customer names from the DataFrame
df_clean.drop(columns=['Customer Name'], inplace=False)

# Printing the 5 first table rows
print(df_clean.head())

### We want to see a visual of profit per State or Province 

# Group by State and sum Profit
state_profit = df_clean.groupby('State or Province')['Profit'].sum().reset_index()
# Set figure size
fig, ax = plt.subplots(figsize=(15, 8))  

# Plot the data on the axes
state_profit.plot(kind='bar', x='State or Province', y='Profit', ax=ax, legend=False, color='blue')

# Customize the plot
ax.set_title('Profit per State or Province')
ax.set_xlabel('State or Province')
ax.set_ylabel('Total Profit')
ax.set_xticklabels(state_profit['State or Province'], rotation=90, ha='right')

# Show the plot
plt.tight_layout()  # Ensure everything fits well
plt.show()
```
